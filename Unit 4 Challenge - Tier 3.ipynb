**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:

```python
# calculate the price ratio 1998 to 2018 price averages for given burough dataframe
# call a more generalized version of the function to do this
def create_price_ratio(borough_df):
    ratio = create_price_ratio_general(borough_df, 1998, 2018)
    return ratio

# calculate the price ratio of given start_yr to given end_yr price averages for given burough dataframe
def create_price_ratio_general(borough_df, start_yr, end_yr):
    start_yr_avg = borough_df[borough_df['Year'] == start_yr]['Average_Price'].mean()
    end_yr_avg = borough_df[borough_df['Year'] == end_yr]['Average_Price'].mean()
    ratio = start_yr_avg / end_yr_avg
    return ratio   
```

```python
# Test create_price_ratio function
create_price_ratio(camden)
```

```python
# Put the London boroughs into a list (use set to avoid duplicates)
London_Boroughs = list(set(properties_clean['London_Borough'].values))
```

```python
London_Boroughs
```

```python
# Create a list of dictionariess of the borough/price_ratio data to allow sorting by price ration
price_ratio_dict_list = []
for borough in London_Boroughs:
    price_ratio_dict = {}
    burough_df = properties_clean[properties_clean['London_Borough'] == borough]
    price_ratio = create_price_ratio(burough_df)
    price_ratio_dict['borough'] = borough
    price_ratio_dict['price_ratio'] = price_ratio    
    price_ratio_dict_list.append(price_ratio_dict)
```

```python
price_ratio_dict_list
```

```python
# Function to help allow a sort by price_ratio
def set_sort_key(dict):
  return dict.get('price_ratio')

# Sort by price_ratio
price_ratio_dict_list.sort(key=set_sort_key)
```

```python
# Sorted price ratio list - the lower the ratio, the higher the increase in average house price
price_ratio_dict_list
```
